```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# input data from Roblox Console
PuristString = "" # delete the quotes and paste the data directly
CCOString = "" # delete the quotes and paste the data directly
timeLimit = 1.5 # adds a limit to length of jumps kept in the data; any jump longer than this is cut
```

```python
# fill purist array

PuristTime = np.empty(len(PuristString))
PuristXZDistance = np.empty(len(PuristString))
PuristYDistance = np.empty(len(PuristString))
PuristTotalDistance = np.empty(len(PuristString))
PuristYawAngle = np.empty(len(PuristString))
PuristPitchAngle = np.empty(len(PuristString))
PuristTotalAngle = np.empty(len(PuristString))
PuristJumpNum = np.empty(len(PuristString))
PuristIterator = 0

for n in range(0, len(PuristString)):
    if(PuristString[n][6] < timeLimit):
        PuristTime[PuristIterator] = np.array(PuristString[n][6])
        PuristXZDistance[PuristIterator] = np.array(PuristString[n][0])
        PuristYDistance[PuristIterator] = np.array(PuristString[n][1])
        PuristTotalDistance[PuristIterator] = np.array(PuristString[n][2])
        PuristYawAngle[PuristIterator] = np.array(PuristString[n][3])
        PuristPitchAngle[PuristIterator] = np.array(PuristString[n][4])
        PuristTotalAngle[PuristIterator] = np.array(PuristString[n][5])
        PuristJumpNum[PuristIterator] = PuristIterator
        PuristIterator += 1
    else:
        print("Cut data at jump number " + str(n))

print(PuristIterator)

# time removal correction
PuristTime = PuristTime[:PuristIterator]
PuristXZDistance = PuristXZDistance[:PuristIterator]
PuristYDistance = PuristYDistance[:PuristIterator]
PuristTotalDistance = PuristTotalDistance[:PuristIterator]
PuristYawAngle = PuristYawAngle[:PuristIterator]
PuristPitchAngle = PuristPitchAngle[:PuristIterator]
PuristTotalAngle = PuristTotalAngle[:PuristIterator]
PuristJumpNum = PuristJumpNum[:PuristIterator]
```

```python
# fill CCO jump array

CCOTime = np.empty(len(CCOString))
CCOXZDistance = np.empty(len(CCOString))
CCOYDistance = np.empty(len(CCOString))
CCOTotalDistance = np.empty(len(CCOString))
CCOYawAngle = np.empty(len(CCOString))
CCOPitchAngle = np.empty(len(CCOString))
CCOTotalAngle = np.empty(len(CCOString))
CCOJumpNum = np.empty(len(CCOString))
CCOIterator = 0

for n in range(0, len(CCOString)):
    if(CCOString[n][6] < timeLimit):
        # if it's above timeLimit, its definitely a mistake, don't count it
        CCOTime[CCOIterator] = np.array(CCOString[n][6])   
        CCOXZDistance[CCOIterator] = np.array(CCOString[n][0])
        CCOYDistance[CCOIterator] = np.array(CCOString[n][1])
        CCOTotalDistance[CCOIterator] = np.array(CCOString[n][2])
        CCOYawAngle[CCOIterator] = np.array(CCOString[n][3])
        CCOPitchAngle[CCOIterator] = np.array(CCOString[n][4])
        CCOTotalAngle[CCOIterator] = np.array(CCOString[n][5])
        CCOJumpNum[CCOIterator] = CCOIterator
        CCOIterator += 1
    else:
        print("Cut data at jump number " + str(n))

print(CCOIterator)

# time removal correction
CCOTime = CCOTime[:CCOIterator]
CCOXZDistance = CCOXZDistance[:CCOIterator]
CCOYDistance = CCOYDistance[:CCOIterator]
CCOTotalDistance = CCOTotalDistance[:CCOIterator]
CCOYawAngle = CCOYawAngle[:CCOIterator]
CCOPitchAngle = CCOPitchAngle[:CCOIterator]
CCOTotalAngle = CCOTotalAngle[:CCOIterator]
CCOJumpNum = CCOJumpNum[:CCOIterator]
```

```python
# printing out averages and standard deviation

print("Stat Results: \n")
print("Average XZ Distance per Jump\nPurist: " + str(PuristXZDistance.mean()) + "\nCCO: " + str(CCOXZDistance.mean()))
print("\n")
print("Average Y Distance per Jump\nPurist: " + str(PuristYDistance.mean()) + "\nCCO: " + str(CCOYDistance.mean()))
print("\n")
print("Average Total Distance per Jump\nPurist: " + str(PuristTotalDistance.mean()) + "\nCCO: " + str(CCOTotalDistance.mean()))
print("\n")
print("Average Yaw Angle per Jump\nPurist: " + str(PuristYawAngle.mean()) + "\nCCO: " + str(CCOYawAngle.mean()))
print("\n")
print("Average Pitch Angle per Jump\nPurist: " + str(PuristPitchAngle.mean()) + "\nCCO: " + str(CCOPitchAngle.mean()))
print("\n")
print("Average Total Angle per Jump\nPurist: " + str(PuristTotalAngle.mean()) + "\nCCO: " + str(CCOTotalAngle.mean()))
print("\n")
print("Average Time per Jump\nPurist: " + str(PuristTime.mean()) + "\nCCO: " + str(CCOTime.mean()))

print("\n")

print("St. Dev. XZ Distance per Jump \nPurist: " + str(PuristXZDistance.std()) + "\nCCO: " + str(CCOXZDistance.std()))
print("\n")
print("St. Dev. Y Distance per Jump \nPurist: " + str(PuristYDistance.std()) + "\nCCO: " + str(CCOYDistance.std()))
print("\n")
print("St. Dev. Total Distance per Jump \nPurist: " + str(PuristTotalDistance.std()) + "\nCCO: " + str(CCOTotalDistance.std()))
print("\n")
print("St. Dev. Yaw Angle per Jump \nPurist: " + str(PuristYawAngle.std()) + "\nCCO: " + str(CCOYawAngle.std()))
print("\n")
print("St. Dev. Pitch Angle per Jump \nPurist: " + str(PuristPitchAngle.std()) + "\nCCO: " + str(CCOPitchAngle.std()))
print("\n")
print("St. Dev. Total Angle per Jump \nPurist: " + str(PuristTotalAngle.std()) + "\nCCO: " + str(CCOTotalAngle.std()))
print("\n")
print("St. Dev. Time per Jump \nPurist: " + str(PuristTime.std()) + "\nCCO: " + str(CCOTime.std()))
print("\n")
```

```python
# Plots (for visualization and the paper)

plt.plot(PuristJumpNum, PuristTime)
plt.title("Tower Purist Airtime")
plt.xlabel("Jump Number")
plt.ylabel("Airtime (s)")
plt.show()

plt.plot(PuristJumpNum, PuristXZDistance)
plt.plot(PuristJumpNum, PuristYDistance)
plt.plot(PuristJumpNum, PuristTotalDistance)
plt.title("Tower Purist Distance")
plt.xlabel("Jump Number")
plt.ylabel("XZ Distance (Studs)")
plt.legend(["XZ Distance","Y Distance","Total Distance"])
plt.show()

plt.plot(PuristJumpNum, PuristYawAngle)
plt.plot(PuristJumpNum, PuristPitchAngle)
plt.plot(PuristJumpNum, PuristTotalAngle)
plt.title("Tower Purist Angle per Jump")
plt.xlabel("Jump Number")
plt.ylabel("Angle (rad)")
plt.legend(["Yaw Angle","Pitch Angle","Total Angle"])
plt.show()

plt.plot(CCOJumpNum, CCOTime)
plt.xlabel("Jump Number")
plt.ylabel("Airtime (s)")
plt.title("Tower CCO Airtime")
plt.show()

plt.plot(CCOJumpNum, CCOXZDistance)
plt.plot(CCOJumpNum, CCOYDistance)
plt.plot(CCOJumpNum, CCOTotalDistance)
plt.title("Tower CCO Distance")
plt.xlabel("Jump Number")
plt.ylabel("XZ Distance (Studs)")
plt.legend(["XZ Distance","Y Distance","Total Distance"])
plt.show()

plt.plot(CCOJumpNum, CCOYawAngle)
plt.plot(CCOJumpNum, CCOPitchAngle)
plt.plot(CCOJumpNum, CCOTotalAngle)
plt.title("Tower Purist Angle per Jump")
plt.xlabel("Jump Number")
plt.ylabel("Angle (rad)")
plt.legend(["Yaw Angle","Pitch Angle","Total Angle"])
plt.show()
```
